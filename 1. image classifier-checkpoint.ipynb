```python
#importing the necessary libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
import os
import cv2
```

```python
#creating two empty list
#image data shape=(n,height,width,rgb)
X=[] #for our images
Y=[] #for our categories


#Creating a folder for pedestrian and one for road
folder_ped='pedestrian'
folder_road='road'



#creating a dictionary which is going to refer the name of the category whether it is a 'padestrian' or 'road' and convert that into a number
name_encode={'pedestrian':0,'road':1}



#function to put the images in the empty list
def image_to_array(folder,name):
    for image in os.listdir(folder):
        loaded_image=Image.open(os.path.join(folder,image))
        resized_image=Image.Image.resize(loaded_image,[100,100])
        
        #first 50 image as a normal image
        image_array=np.array(resized_image)
        X.append(image_array)
        Y.append(name_encode[name])
        
        #another 50 image as flipped image
        #fliping the photo in a opposite direction, if the person is seeing left then it will appear as right
        image_flipped=cv2.flip(image_array,1)
        X.append(image_flipped)
        Y.append(name_encode[name])
        
        #another 50 image as blured image
        #adding blur effect to our photos
        image_blurred=cv2.blur(image_array,(2,2))
        X.append(image_blurred)
        Y.append(name_encode[name])
        
        #another 50 image as flipped as well as blured image
        #fliping the image as well as bluring the image
        image_flipped_blur=cv2.blur(image_flipped,(2,2))
        X.append(image_flipped_blur)
        Y.append(name_encode[name])
        
        #now we have 200 images
        
#function to show the image        
def show_image(index):
    plt.imshow(X[index])
    plt.show()
    print(Y[index])
    
    

#calling the function to put the image data into the list
image_to_array(folder_ped,'pedestrian')
image_to_array(folder_road,'road')



#[1,0]means pedestrian
#[0,1]means road
Y=to_categorical(Y,num_classes=2)



#cast all the data into numpy array. And the range will be from -1 to +1
X=(np.array(X)-127.5)/127.5


show_image(40)#image of a pedestrian
show_image(60)#image of a road
show_image(150)#image blured
show_image(250)
```

```python
#creating the architecture of our neural network
from keras.layers import Activation, Conv2D, Dense, Flatten, MaxPool2D, BatchNormalization, Dropout
#sigmoid function is available in Activation module
#conv2D is convulational layers which have kernals

from keras.models import Sequential



#initializing our model
model=Sequential()



#adding the inner layer in our models
model.add(Conv2D(32,(5,5),padding='same',activation='relu',input_shape=(100,100,3)))
model.add(BatchNormalization())
#32 nodes in our first layer
#(5,5) is our stride

#adding pooling in our layer
model.add(MaxPool2D(pool_size=(2,2)))


#second convulation layer and second maxpooling layer
model.add(Conv2D(100,(5,5),padding='same',activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(rate=0.5))

model.add(MaxPool2D(pool_size=(2,2)))

#third convulation layer and third maxpooling layer
model.add(Conv2D(200,(5,5),padding='same',activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(rate=0.5))


model.add(MaxPool2D(pool_size=(2,2)))


#flattern our data
model.add(Flatten())

model.add(Dense(124))
model.add(Activation('relu'))
model.add(Dense(2))
model.add(Activation('sigmoid'))
model.summary()
```

```python
#spliting the data into train and test
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.1)

print(Y[60])
```

```python
#setting up an optimizer
from tensorflow.keras.optimizers import Adam

optimizer=Adam(learning_rate=0.001) #lr stands for Learning rate



#compiling our model
model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])



#training our model
h=model.fit(X_train,y_train,batch_size=10,epochs=20,validation_data=(X_test,y_test))

#saving our model
model.save('road_pedestrian_CNN.h5')


#visulazing the accuracy with epochs
plt.plot(h.history['accuracy'])
plt.plot(h.history['val_accuracy'])
plt.title('CNN accuracy- Train and Test')
plt.xlabel("Epoch number")
plt.ylabel("Accuracy")
plt.legend(loc='lower right')
```

```python
#running predictions
from keras.models import load_model

model=load_model('road_pedestrian_CNN.h5')

def prediction(index_number):
    img=(np.array(X[index_number])-127.5)/127.5
    img=img.reshape(1,100,100,3)
    prediction=np.argmax(model.predict(img), axis=-1)
    print(prediction)
    if prediction[0]==0:
        print("Prediction: Pedestian")
    elif prediction[0]==1:
        print("Prediction: Road")
    else:
        print("Invalid prediction")
    plt.imshow(X[index_number])
    plt.show()

    
prediction(200)
```
